```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 讀取兩份 CSV 檔案
version1 = input("version1:")
version2 = input("version2:")
df1 = pd.read_csv(f'{version1}/training_cruve.csv')
df2 = pd.read_csv(f'{version2}/training_cruve.csv')

# 2. 設定圖表風格與大小
plt.figure(figsize=(10, 6))
plt.style.use('seaborn-v0_8-whitegrid') 

# 3. 定義繪製曲線與陰影的函數
def plot_rl_curve(df, label, color):
    x = df.index
    plt.plot(x, df['mean'], label=label, color=color, linewidth=2)
    plt.fill_between(x, df['min'], df['max'], color=color, alpha=0.2)

# 4. 畫出兩組實驗數據
plot_rl_curve(df1, label=f'Training Curve {version1}', color='#1f77b4')  
plot_rl_curve(df2, label=f'Training Curve {version2}', color='#ff7f0e')  

# ---------------------------------------------------
# 5. 指定 Y 軸的顯示範圍 (下限, 上限)
# 你可以把 -2000 和 1000 改成你想要看的最精確範圍
plt.ylim(-600, 600) 
# ---------------------------------------------------

# 6. 設定標題與軸標籤
plt.title('SAC Agent Training Performance', fontsize=14, pad=15)
plt.xlabel('Episodes / Training Steps', fontsize=12)
plt.ylabel('Reward', fontsize=12)

# 7. 加上圖例並優化排版
plt.legend(loc='lower right', fontsize=11)
plt.tight_layout()

# 8. 儲存圖表並顯示
plt.savefig('sac_training_comparison_ylim.png', dpi=300, bbox_inches='tight')
plt.show()
```
